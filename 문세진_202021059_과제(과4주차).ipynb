### 반복문을 사용해서 '음식' 이미지 200장 수집

```python
encText = urllib.parse.quote("음식")

for i in range(1, 11):
    url = "https://openapi.naver.com/v1/search/image?query=" + encText  + "&display=20&start=" + str(i)
    request = urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id",client_id)
    request.add_header("X-Naver-Client-Secret",client_secret)
    response = urllib.request.urlopen(request)
    rescode = response.getcode()
    if(rescode==200):
        response_body = response.read()
        result = response_body.decode('utf-8')
        images_result = json.loads(result)
        images_list.extend(images_result['items'])  # 결과를 리스트에 추가.
    else:
        print("Error Code:" + rescode)
```

```python
for i in images_list:
    print(i) #추출한 데이터 텍스트 형식으로 출력
```

## 네이버 블로그, 뉴스, 책 등에서 원하는 키워드를 바탕으로 1,000개 이상의 검색 데이터를 수집

```python
news_list = []
```

```python
encText = urllib.parse.quote("손흥민")

for i in range(1, 11):
    url = "https://openapi.naver.com/v1/search/news?query=" + encText  + "&display=100&start=" + str(i)
    request = urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id",client_id)
    request.add_header("X-Naver-Client-Secret",client_secret)
    response = urllib.request.urlopen(request)
    rescode = response.getcode()
    if(rescode==200):
        response_body = response.read()
        result = response_body.decode('utf-8')
        news_result = json.loads(result)
        news_list.extend(news_result['items'])  # 결과를 리스트에 추가.
    else:
        print("Error Code:" + rescode)
```

```python
for i in news_list:
    print(i) #추출한 데이터 텍스트 형식으로 출력
```

```python
df_news = pd.DataFrame(news_list)
```
