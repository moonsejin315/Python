```python
import pandas as pd
import numpy as np

df = pd.read_csv("/Users/sejinmoon/Desktop/24-1/data_science_hw/StudentsPerformance.csv")
df.head()
```

# 전체 학생들의 math, reading, writing 점수의 평균

```python
math_score = df['math score'].mean()
reading_score = df['reading score'].mean()
writing_socre = df['writing score'].mean()

print(f"math 평균 : {math_score}")
print(f"reading 평균 : {reading_score}")
print(f"writing 평균 : {writing_socre}")
```

# Group별 math, reading, writing 점수의 평균

```python
group_math = df.groupby('race/ethnicity')['math score'].mean()
group_math
```

```python
group_reading = df.groupby('race/ethnicity')['reading score'].mean()
group_reading
```

```python
group_writing = df.groupby('race/ethnicity')['writing score'].mean()
group_writing
```

# Group C에 속하는 남자와 여자들 각각의 math, reading, writing 점수의 평균을 구하세요.

```python
group_c = df.groupby('race/ethnicity').get_group('group C')
c_male = group_c.groupby('gender').get_group('male')
c_female = group_c.groupby('gender').get_group('female')

male_math = c_male['math score'].mean()
male_reading = c_male['reading score'].mean()
male_writing = c_male['writing score'].mean()

female_math = c_female['math score'].mean()
female_reading = c_female['reading score'].mean()
female_writing = c_female['writing score'].mean()

print(f"남자 -> math : {male_math}, reading : {male_reading}, writing : {male_writing}")
print(f"여자 -> math : {female_math}, reading : {female_reading}, writing : {female_writing}")
```

# 남자 중 수학 점수가 80점 이상인 학생의 수

```python
male_math = df.groupby('gender').get_group('male')
result = male_math['math score']>=80
print(f"{result.count()}명")
```

# 시험 준비 과정을 완료 한 학생들만 남기고 모두 삭제!

```python
none_students = df.groupby('test preparation course').get_group('none')
completed_students = df.drop(none_students.index, axis = 0)
completed_students
```
